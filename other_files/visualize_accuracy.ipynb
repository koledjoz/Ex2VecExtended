```python
import matplotlib.pyplot as plt

import numpy as np
```

```python
extended = np.loadtxt('extended_accuracy.txt')
original = np.loadtxt('original_accuracy.txt')
```

```python
plt.figure(figsize=(16, 8))

plt.plot(extended, label='Extended')
plt.plot(original, label='Original')

plt.legend(fontsize=14)

plt.title('Plot of accuracy@N', fontsize=18)  # Set title and its font size
plt.xlabel('N', fontsize=16)  # X-axis label
plt.ylabel('Accuracy@N', fontsize=16)  # Y-axis label

plt.grid()

plt.ylim((-0.05, 0.55))

plt.tick_params(axis='both', labelsize=14)

plt.savefig('accuracy_plot.png', dpi=300, bbox_inches='tight')
```
